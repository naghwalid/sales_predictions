```python
# 2) Datatypes of each variable
data_types = df.dtypes
print( data_types)
```

```python
# 3) Drop duplicates (if any)
duplicates=df.duplicated()
if duplicates.any():
   df = df.drop_duplicates()
   print('Dropped',duplicates.sum(),'duplicate rows')
```

```python
# 4) Identify missing values
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
# 5) Address missing values
# For numeric columns (e.g., item_weight), you can replace missing values with the mean
df['Item_Weight'].value_counts()
df['Item_Weight'].fillna(value =df['Item_Weight'].mean(),inplace=True )
df
```

```python
# For categorical columns (e.g., outlet_size), you can replace missing values with the mode
df['Outlet_Size'].value_counts()
# df['Outlet_Size'].dropna(inplace=True)
df['Outlet_Size'].fillna(value='Medium',inplace=True)
df
```

```python
# 6) Confirm no missing values after addressing them
missing_values_after = df.isnull().sum()
print("Missing values after addressing them:", missing_values_after)
```

```python
#Fix inconsistent categories of data
#let's assume we want to fix inconsistent category names in ' item_fat_content' column
df['Item_Fat_Content']= df['Item_Fat_Content'].replace({'LF':'Low Fat','reg':'Regular'})
df['Item_Fat_Content'].value_counts()
```
