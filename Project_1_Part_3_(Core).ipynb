```python
import matplotlib.pyplot as plt
import seaborn as sns
#1) Bar Plot: Visualizing the count of each outlet size category.
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Outlet_Size')
plt.title('Count of Outlet Sizes')
plt.xlabel('Outlet Size')
plt.ylabel('Count')
plt.show()
```

```python
#2) Scatter Plot: Examining the relationship between item weight and item visibility
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Item_Weight', y='Item_Visibility')
plt.title('Item Weight vs Item Visibility')
plt.xlabel('Item Weight')
plt.ylabel('Item Visibility')
plt.show()
```

```python
# Histogram of Item Weight Distribution:
plt.figure(figsize=(10,6))
sns.histplot(data=df, x='Item_Weight', stat='count', bins='auto')
plt.xlabel('Item Weight')
plt.ylabel('Count')
plt.title('Distribution of Item Weight')
plt.show()
```

```python
# Boxplot of Item Fat Content and Item Weight:

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='Item_Fat_Content', y='Item_Weight')
plt.xlabel('Item Fat Content')
plt.ylabel('Item Weight')
plt.title('Relationship between Item Fat Content and Item Weight')
plt.show()
```

```python
sns.boxplot(data=df, x='Outlet_Size', y='Item_Visibility')
plt.title("Item Visibility by Outlet Size")
plt.xlabel("Outlet Size")
plt.ylabel("Item Visibility")
plt.show()

```
