```python
# Importing the required libraries
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df['Item_Weight'].head()
```

```python
# Histograms
sns.histplot(data=df, x='Item_Weight', y=None, hue=None,  stat='count', bins='auto')
plt.xlabel('Item Weight')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data=df, x='Item_Visibility', y=None, hue=None,  stat='count', bins='auto')
plt.xlabel('Item Visibility')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(data=df, x='Item_Outlet_Sales', y=None, hue=None,  stat='count', bins='auto')
#<Axes: xlabel='Item_Outlet_Sales', ylabel='Count'>
```

```python
# Boxplots
sns.boxplot(data=df, x='Item_Fat_Content', y='Item_Weight')
```

```python
sns.boxplot(data=df, x='Outlet_Size', y='Item_Weight')
plt.xlabel('Outlet Size')
plt.ylabel('Item Weight')
plt.title('Boxplot of Item Weight by Outlet Size')
plt.show()
```

```python
# Correlation heatmap
cor = df.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(cor, cmap='plasma', annot=True)
plt.title('Correlation between Numeric Features')
plt.show()
```
