```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/drive/MyDrive/Dataset/sales_predictions_2023.csv')
```

```python
#Dropping duplicates :
df.drop_duplicates(inplace=True)
```

```python
#Fix inconsistencies in categorical data:
df['Item_Fat_Content']=df['Item_Fat_Content'].replace({'LF':'Low','reg':'Regular','Low fat':'Low Fat'})
```

```python
X = df.drop(columns=['Item_Outlet_Sales'])
y = df['Item_Outlet_Sales']
X.head()
```

```python
# Perform a train-test split:
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a SimpleImputer object to handle missing values
imputer = SimpleImputer(strategy='mean')

# Apply imputation on the numerical features in train and test sets
X_train['Item_Weight'] = imputer.fit_transform(X_train['Item_Weight'].values.reshape(-1, 1))
X_test['Item_Weight'] = imputer.transform(X_test['Item_Weight'].values.reshape(-1, 1))

# Create a StandardScaler object to scale the numerical features
scaler = StandardScaler()
```
